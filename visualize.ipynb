```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getAverage(data, className):
    return np.mean(
        [record[className] for record in data ],
        axis=0
    )
```

```python
data = [
    json.loads(text)
    for text in open('data/train.json')
    
]
classNames= ['reactivity', 'deg_pH10', 'deg_Mg_pH10', 'deg_50C', 'deg_Mg_50C']
averages = {
    className: getAverage(data, className)
    for className in classNames
}
```

```python
for className in classNames:
    plt.plot(averages[className], label=className)
plt.legend()
plt.title('Reactivies For Each Class')
```
